Extaracted Data_csv file we will us the Pandas to extract the clear data and usage of the of the clean data to form better visualization of the Data to make it in better in Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

medium_data_csv = pd.read_csv('Extracted_Data.csv')
medium_data_csv.head(4)
```

To find out the Higest Publisher name from the csv file from the company bases

```python
import matplotlib.pyplot as plt

# Calculate value counts for 'Publisher'
total = medium_data_csv['Publisher'].str.replace('in',"",regex=False).value_counts()

# Calculate the total sum
total_sum = total.sum()

# Calculate the percentage for each publisher
per = [(each / total_sum) * 100 for each in total]

# Print each publisher and its percentage
for publisher, percentage in zip(total.index, per):
    print(f'{publisher}: {percentage:.2f}%')

# Create a pie chart
explode =[0.1,0.1,0.1,0.2,0.1,0.2,0.1]
plt.pie(per, autopct='%.1f%%',explode=explode, labels=total.index)
# plt.figure(figsize=(5, 8))
plt.title('Medium Site Data')
plt.axis('equal')
plt.show()
```
